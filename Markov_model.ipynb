```python
with open('edgar_allan_poe.txt') as f:
    lines = f.readlines()
with open('robert_frost.txt') as f:
    lines2 = f.readlines()
```

```python
lines
```

```python

text1=[]
text2=[]
for line in lines:
  text1.append(re.sub(r"(@\[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", line).lower())  
while('' in  text1 ):
  text1.remove('')
for line in lines2:
  text2.append(re.sub(r"(@\[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", line).lower())  
while('' in  text2 ):
  text2.remove('')
text2
```

```python
#for 1st poem - code to process the text file poem into a datafram and label them accordingly to their author
from itertools import zip_longest
d1=zip_longest(text1,['0'],fillvalue='0')
d1=dict(d1)
inv_map = {}
for k, v in d1.items():
    inv_map[v] = inv_map.get(v, []) + [k]
inv_map
df1=pd.DataFrame.from_dict(data=inv_map)

df1['label']='0'
df1.rename(columns = {'0':'poem'}, inplace = True)
df1
#for 2nd poem
from itertools import zip_longest
d2=zip_longest(text2,['1'],fillvalue='1')
d2=dict(d2)
inv_map = {}
for k, v in d2.items():
    inv_map[v] = inv_map.get(v, []) + [k]
inv_map
df2=pd.DataFrame.from_dict(data=inv_map)

df2['label']='1'
df2.rename(columns = {'1':'poem'}, inplace = True)
df2
```

```python
df=pd.concat([df1,df2])
X=df['poem']
Y=df['label']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
print(len(y_train))
print(x_train[:5])
len(y_test)
```

```python
# create mapping from unique word to word
idx=1
word2idx={'<unk>':0}
for text in x_train:
  tokens=text.split()
  for token in tokens:
    if token not in word2idx.values():  #.values( ) cz aparently in fuction aint checking values anymore sadge
      word2idx[idx]=token
      idx+=1
word2idx={v:k for k,v in word2idx.items()}
len(word2idx)
# if "and" not in word2idx.values():
#   print('yes')
# else:
#   print('no')
```

```python
x_train_int=[]
x_test_int=[]

for text in x_train:
  tokens=text.split()
  # print(token for token in tokens) this ones wrong but the bottoms ones correct somehow eh??
  line_as_int=[word2idx[token] for token in tokens]
  x_train_int.append(line_as_int)

for text in x_test:
  tokens=text.split()
  line_as_int=[word2idx.get(token,0) for token in tokens]
  x_test_int.append(line_as_int)

print(x_train_int[:5])
# k=[word2idx[token] for token in tokens]
# k
```

```python
def compute_counts(text_int,A,pi):
  for tokens in text_int:
    last_idx=None
    for idx in tokens:
      if last_idx is None:#first entry moment
        pi[idx]+=1
      else:
        A[last_idx,idx]+=1
      last_idx=idx

compute_counts([t for t,y in zip(x_train_int,y_train) if y=='0'],A0,pi0)   #lmao the y==0 was actually a string not a int sadge
compute_counts([t for t,y in zip(x_test_int,y_test) if y=='1'],A1,pi1)  
```
